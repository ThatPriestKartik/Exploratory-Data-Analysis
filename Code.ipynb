```python
#lets get a clean athelete age 
```

```python
df2['athlete_age']=2020 - df['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split().str.get(0)
```

```python
df2.isna().sum()
```

```python
df3.shape
```

```python
#checking duplicates
```

```python
df3[df3.duplicated() == True]
```

```python
#good, now lets just reset index
df3.reset_index(drop = True)
```

```python
df3.info()
```

```python
#lets make age, avg speed as int
df3['athlete_age']=df3['athlete_age'].astype(int)
```

```python
df3.info()
```

```python
df3['Athlete average speed']=df3['Athlete average speed'].astype(float)
```

```python
df3.info()
```

```python
#lets rename these column names
df3=df3.rename(columns={'Year of event': 'YEAR' ,
'Event dates': 'DATE',
'Event name': 'RUN NAME',
'Event distance/length': 'RUN DIST',
'Event number of finishers': 'NUMBER OF FINISHERS',
'Athlete performance': 'PERFORMANCE',
'Athlete gender': 'GENDER',
'Athlete average speed': 'AVG SPEED',
'Athlete ID': 'ATHLETE ID',
'athlete_age': 'AGE'})
```

```python
df3.head(113)
```

```python
sns.lmplot(df3,x='AGE',y='AVG SPEED', hue= 'GENDER')
```
